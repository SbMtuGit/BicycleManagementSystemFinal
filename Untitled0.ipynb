```python
import numpy as np
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
import librosa
import librosa.display
#
```

```python
filename = "DNB_BREAK_14.wav"
Y, sr = librosa.load(filename)
num_Samples= len(Y)
print(num_Samples)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(Y)
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.title('Audio waveform')

plt.show()
```

```python
## extract the pitch using the normalize pitch method
pitch=librosa.util.normalize(Y)
plt.figure(figsize=(10, 4))
plt.plot(pitch)
plt.xlabel('Time(Samples)')
plt.ylabel('Pitch (Hz)')
plt.title('Pitch Contor')
plt.show()
```

```python
# Extract Mel-Frequency Cepstral Coefficients (MFCCs)
mfccs = librosa.feature.mfcc(y=Y, sr=sr, n_mfcc=13)

# Plot the MFCCs
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, sr=sr, x_axis='time', y_axis='mel', cmap='viridis')
plt.colorbar(format='%0.1f dB')
plt.xlabel('Time')
plt.ylabel('Mel Frequency')
plt.title('MFCCs')
plt.show()
```

```python
# Extract tempo and beat times
tempo, beat_times = librosa.beat.beat_track(y=Y, sr=sr)

# Print the tempo and beat times
print("Tempo:", tempo)
print("Beat times:", beat_times)
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load the audio file
y, sr = librosa.load("DNB_BREAK_14.wav")

# Extract pitch using the PYIN algorithm
pitches, magnitudes = librosa.piptrack(y=y, sr=sr)

# Get the time values for the pitch data
times = librosa.times_like(pitches)

# Extract MFCCs
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)  # Adjust n_mfcc as needed

# Get the time values for the MFCC data
mfcc_times = librosa.times_like(mfccs)

# Choose a specific MFCC coefficient (e.g., the first one)
mfcc_coeff = mfccs[0]

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(mfcc_coeff, pitches[0], alpha=0.5)  # Using the first pitch track
plt.xlabel("MFCC Coefficient")
plt.ylabel("Pitch (Hz)")
plt.title("Scatter Plot of Pitch vs. MFCC")
plt.grid(True)
plt.show()
```
